```python
import numpy as np
import pandas as pd
import seaborn as sns
from scipy import stats
import matplotlib.pyplot as plt
from sklearn.svm import SVC
```

```python
from sklearn.datasets import load_breast_cancer
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
```

```python
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
cancer = load_breast_cancer()
```

```python
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, random_state=1)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
X_train
```

```python
print(y_train.shape)
print(y_test.shape)
```

```python
mmscaler = MinMaxScaler()
```

```python
mmscaler.fit(X_train)
```

```python
X_train_scaled = mmscaler.transform(X_train)
X_train_scaled
```

```python
X_test_scaled = mmscaler.transform(X_test)
X_test_scaled
```

```python
mmsvm = SVC(100)
mmsvm.fit(X_train_scaled, y_train)
print("테스트 세트 정확도: {:.2F}".format(mmsvm.score(X_test_scaled, y_test)))
```

```python
sdscaler = StandardScaler()
sdscaler.fit(X_train)
X_train_sdscaled = sdscaler.transform(X_train)
X_test_sdscaled = sdscaler.transform(X_test)
```

```python
sdsvm = SVC(100)
sdsvm.fit(X_train_sdscaled, y_train)
print("테스트 세트 정확도: {:.2F}".format(sdsvm.score(X_test_sdscaled, y_test)))
```
